```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=1,n_redundant=0)
```

```python
X
```

```python
y
```

```python
X_temp = pd.DataFrame(X)
```

```python
plt.scatter(X_temp[0],X_temp[1])
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
model = SVC(kernel = 'poly')
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score , classification_report
```

```python
acc = accuracy_score(y_test,y_pred)
cm = confusion_matrix(y_test,y_pred)
report = classification_report(y_test,y_pred)
```

```python
print(acc)
print(cm)
print(report)
```

## Find Out The Best Parameter For SVC

```python
from sklearn.model_selection import GridSearchCV # It Iterates all Paratmeters we given based on that which one best param for that data set
```

```python
param_grid = {
    'C':[0.1,1,0.01,10,100,1000],
    'gamma' : [1,0.1,0.01,0.001,0.0001],
    'kernel':['rbf','sigmoid','poly','linear']
}
```

```python
grid = GridSearchCV(SVC(),param_grid=param_grid,refit=True,cv=5,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
#This Is the Best Parameter For This Data Set
model = SVC(C=10, gamma=1, kernel='linear')
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
acc = accuracy_score(y_test,y_pred)
cm = confusion_matrix(y_test,y_pred)
report = classification_report(y_test,y_pred)
```

```python
print(acc)
print(cm)
print(report)
```
