```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
```

```python
plt.plot(x,y,color='red',linestyle='--',marker='*',linewidth=3,markersize=9)
plt.title("Line plot")
plt.xlabel("Value")
plt.grid(True)
plt.ylabel('Squered Value')
```

```python
# Multiple Plots (Subplots)
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5))

plt.subplot(1,2,1)  # Intha (1,2,1) -> Ennathuna 1X2 nu oru grid create pannum athula 3rd paramter entha index la plot pannaum nu namma kudulam
plt.plot(x,y1)
plt.title('First Plot')

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title('Second Plot')

plt.show()
```

```python
# Bar Plots
categories = ['A','B','C']
values = [1,2,3]
 
plt.bar(categories,values)
```

```python
# Histogram
data = [1,1,2,2,2,3,3,3,3]
plt.hist(data,bins=3,color='orange',edgecolor='black')
```

```python
# Craeting a Scatter Plot
plt.scatter(x,y,marker='*')
```

```python
# Creating a PIE chart
labels = ['a','b','c','d','e']
sizes = [10,50,20,15,5]
explode = (0.2,0,0,0,0)  # It split the selecteed slice from the pie chart
plt.pie(sizes,labels=labels,explode=explode,shadow=True)
```
