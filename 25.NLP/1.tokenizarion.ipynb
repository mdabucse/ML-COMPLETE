```python
corpus = """Hello Weolcome, to Abu Classes.
Please Watch entire course! to become expert in NLP.
"""
```

```python
print(corpus)
```

```python
# Sentence into Paragraph
from nltk.tokenize import sent_tokenize
```

```python
import nltk
nltk.download('all')
```

```python
document = sent_tokenize(corpus)
```

```python
document
```

```python
# Sentence to Word Tokenizezation
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
# Based on the Punctuations Sentence Splitting
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

```python
# Tree Bank Tokenizer -> This is gives the Words with Punctuation splitting
# Normel methods -> This is Good.Stuff. -> ['This','is','Good','.','Stuff','.']
# Tree Bank Tokenizer -> This is Good. -> ['This','is','Good.','Stuff','.'] The Last Dot only treated as seperated word
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer = TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
