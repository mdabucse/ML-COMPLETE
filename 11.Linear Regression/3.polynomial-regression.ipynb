```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = 6 * np.random.rand(100,1) - 3
y = 0.5 * x**2 + 1.5*x+2+np.random.randn(100,1)
```

```python
# Plot the values
plt.scatter(x,y)
```

```python
# For this Create a Simple Linear Regression How it is Perform and See the Accuracy
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
linear_regression = LinearRegression()
linear_regression.fit(X_train,y_train)
```

```python
# Calculate The Accuracy Of The Model
from sklearn.metrics import r2_score
score = r2_score(y_train,linear_regression.predict(X_train))
print('The Accuracy for The Simple Linear Regression is ',score*100)
```

```python
plt.scatter(X_train,y_train)
plt.plot(X_train,linear_regression.predict(X_train),color='r')
```

```python
# Now Check the Polynomial Regression Of The Line
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2,include_bias=True)
```

```python
X_train = poly.fit_transform(X_train)
y_train = poly.transform(y_train)
```

```python
polnynomial_regression = LinearRegression()
```

```python
polnynomial_regression.fit(X_train,y_train)
```

```python
plt.scatter(X_train,y_train)
plt.scatter(X_train,polnynomial_regression.predict(X_train),color='r')
```

```python
score = r2_score(y_train,polnynomial_regression.predict(X_train))
print('The Accuracy Of the Model Using Polynomial Regression is ',score*100)
```
